```python
from random import randint 
player = int(input("How many players are bowling today? "))
names =[]
for i in range(1,player+1):
    i = input("Please input player's name? ")
    names.append(i)
    
bowling = open("bowling.txt", "w")
def bowling_game(names):
    for k in range(1,11):
        for j in names:
            score = randint(0,10)
            score2 = randint(0,10-score)
            if score == 10: #Strike
                print(j, "Round ", k)
                print(score)
                bowling.write(str(j) + '\n')
                bowling.write(str(score) + '\n')
                bowling.write(str(0) + '\n')
            else:
                print(j, "Round ", k)
                print(score, score2)
                bowling.write(str(j) + '\n')
                bowling.write(str(score) + '\n')
                bowling.write(str(score2)+ '\n')
bowling_game(names)
bowling.close()

def finalScore(player):
    with open("bowling.txt", 'r') as f:
        total=0
        if player in names:
            for line in f:
                if player in line:                
                    roll1=int(f.readline().rstrip())
                    roll2=int(f.readline().rstrip())
                    total+=roll1+roll2
                else:
                    pass
        return total

    f.close()

for i in names:
    print(i, "Final score: ", finalScore(i))
```
